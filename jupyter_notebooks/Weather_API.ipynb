### Hamburg

#### The Weather Data for Hamburg

```python
# List of the Latitude and Longetude Hamburg
lat_HH = [53.550556]
lon_HH = [9.993333]

# Empty dataframe where our weather data about every city comes in
weather_HH = pd.DataFrame([])

# Start and end date of the period being analyzed
start = datetime(2017, 1, 1)
end = datetime(2021, 12, 31)

# For loop to add the lat and lon from each city via zip command to get our dataframe
for l1, l2 in zip(lat_HH, lon_HH):
    city_point_HH = Point(l1, l2)

#In the loop to get daily data for each city
    weather_HH_data = Daily(city_point_HH, start, end)
#In the loop to query the data from the api    
    weather_HH_data = weather_HH_data.fetch()
#In the loop to build new colum for Latitude and Longetude each to get the values
    weather_HH_data['lat'] = l1
    weather_HH_data['lon'] = l2
# Reset the index
    weather_HH_data.reset_index(level = 0, inplace=True) 


#Append the values to our final Dataframe
    weather_HH = weather_HH.append(weather_HH_data)

# Dataframe about weather data in each city
weather_HH
```

#### Aggregated weather columns Hamburg

##### Average temperature for each month from 2017 until the end of 2021

```python
# Groupby date and set it to a month period then get the avg. temp
weather_HH_t = weather_HH.groupby(pd.PeriodIndex(weather_HH['time'], freq="M"))['tavg'].mean()


# put the series above in a dataframe
weather_HH_temp = pd.DataFrame(weather_HH_t)

# add the lat and lon to the new table for joining later with the other aggregated weather functions
weather_HH_temp['lat'] = 53.550556
weather_HH_temp['lon'] = 9.993333
weather_HH_temp['city'] = 'hamburg'

# reset the index
weather_HH_temp = weather_HH_temp.reset_index()
```

```python
# Rename columns
weather_HH_temp.rename(columns = {"time":"date"}, inplace=True)
```

```python
# print all for test 
weather_HH_temp
```

##### Average sunmin for each month from 2017 until the end of 2021

```python
# Groupby date and set it to a month period then get the avg. sunhours
weather_HH_s = weather_HH.groupby(pd.PeriodIndex(weather_HH['time'], freq="M"))['tsun'].mean()

# put the series above in a dataframe
weather_HH_sunh = pd.DataFrame(weather_HH_s)

# reset the index
weather_HH_sunh = weather_HH_sunh.reset_index()
```

```python
# Rename columns
weather_HH_sunh.rename(columns = {"time":"date"}, inplace=True)
```

```python
weather_HH_sunh 
```

##### Count the monthly Raindays

```python
# Filter the raindays and the not raindays
# Apply a function along an axis of the DataFrame
weather_HH["rain_days"] = weather_HH['prcp'].apply(lambda x: 'rainy_day' if x >= 2.5 else 'not_rainy_day')
```

```python
# New variable with only the raindays
weather_HH_rainy = weather_HH[weather_HH["rain_days"]=='rainy_day']

# Group all rain days based on month from 2017 until 2021 and count all these days in a month
weather_HH_rainy = weather_HH_rainy.groupby(pd.PeriodIndex(weather_HH_rainy['time'], freq="M"))['rain_days'].value_counts()

# Put the new variable in a Dataframe
weather_HH_rainy = pd.DataFrame(weather_HH_rainy)

# drop the column level(s) - default is level=0 (We got here twice raindays but we only need the count raindays column so we drop the level= 0 column )
weather_HH_rainy = weather_HH_rainy.droplevel("rain_days")
weather_HH_rainy = weather_HH_rainy.reset_index()
```

```python
# Rename columns
weather_HH_rainy.rename(columns = {"time":"date"}, inplace=True)
```

```python
# print all for test 
weather_HH_rainy
```

```python
# New variable with only days without rain
weather_HH_notrain = weather_HH[weather_HH["rain_days"]=='not_rainy_day']

# Group all not rain days based on month from 2017 until 2021 and count all these days in a month
weather_HH_notrain = weather_HH_notrain.groupby(pd.PeriodIndex(weather_HH_notrain['time'], freq="M"))['rain_days'].value_counts()

# Put the new variable in a Dataframe
weather_HH_notrain = pd.DataFrame(weather_HH_notrain)

# drop the column level(s) - default is level=0 (We got here twice raindays but we only need the count raindays column so we drop the level= 0 column )
weather_HH_notrain = weather_HH_notrain.droplevel("rain_days")
weather_HH_notrain = weather_HH_notrain.reset_index()
```

```python
# Rename columns
weather_HH_notrain.rename(columns = {"time":"date","rain_days":"day_without_rain"}, inplace=True)
```

```python
# print for test
weather_HH_notrain
```

```python
# Filter the snowdays and the not snowdays
# Apply: Apply a function along an axis of the DataFrame.
weather_HH["snow_days"] = weather_HH['snow'].apply(lambda x: 'snow_day' if x > 0 else 'not_snow_day')
```

```python
# New variable with only days with snow
weather_HH_snowday = weather_HH[weather_HH["snow_days"]=='snow_day']

# Group all snowdays based on month from 2017 until 2021 and count all these days in a month
weather_HH_snowday = weather_HH_snowday.groupby(pd.PeriodIndex(weather_HH_snowday['time'], freq="M"))['snow_days'].value_counts()

# Put the new variable in a Dataframe
weather_HH_snowday = pd.DataFrame(weather_HH_snowday)

# drop the column level(s) - default is level=0 (We got here twice raindays but we only need the count raindays column so we drop the level= 0 column )
weather_HH_snowday = weather_HH_snowday.droplevel("snow_days")
weather_HH_snowday = weather_HH_snowday.reset_index()
```

```python
weather_HH_snowday.rename(columns = {"time":"date"}, inplace=True)
```

```python
# New variable with only days with snow
weather_HH_notsnowday = weather_HH[weather_HH["snow_days"]=='not_snow_day']

# Group all snowdays based on month from 2017 until 2021 and count all these days in a month
weather_HH_notsnowday = weather_HH_notsnowday.groupby(pd.PeriodIndex(weather_HH_notsnowday['time'], freq="M"))['snow_days'].value_counts()

# Put the new variable in a Dataframe
weather_HH_notsnowday = pd.DataFrame(weather_HH_notsnowday)

# drop the column level(s) - default is level=0 (We got here twice raindays but we only need the count raindays column so we drop the level= 0 column )
weather_HH_notsnowday = weather_HH_notsnowday.droplevel("snow_days")
weather_HH_notsnowday = weather_HH_notsnowday.reset_index()
```

##### Merge all tables together

```python
#merge all tables on date
df_weather_HH = weather_HH_temp.merge(weather_HH_sunh, how='left' , on='date').merge(weather_HH_rainy,how='left', on='date').merge(weather_HH_notrain,how='left', on='date').merge(weather_HH_snowday,how='left',on="date").merge(weather_HH_notsnowday,how='left',on='date')
```

##### Last cleaning of the table

```python
# Change the order
df_weather_HH = df_weather_HH[["date","tavg","tsun","rain_days","day_without_rain","snow_days","days_without_snow","city","lat","lon"]]
```

```python
# To see the Datatypes that had to change (date to datetime, rain_days and snow_days to int as well as fill the NaN with 0)
df_weather_HH.info()
```

```python
# clean the raindays column
# fill all nan with 0 to show that months on which the event did not take place are 0 
df_weather_HH["rain_days"] = df_weather_HH["rain_days"].fillna(0)

# Change it to an int
df_weather_HH["rain_days"] = df_weather_HH["rain_days"].astype('int')

# clean the snowdays column
# fill all nan with 0 to show that months on which the event did not take place are 0 
df_weather_HH["snow_days"] = df_weather_HH["snow_days"].fillna(0)

# Change it to an integer
df_weather_HH["snow_days"] = df_weather_HH["snow_days"].astype('int')
```

```python
# To change the date from a period[M] to a datetime64[ns] 
df_weather_HH = df_weather_HH.set_index("date").to_timestamp().reset_index()
```

```python
# Check if everything fits now
display(df_weather_HH.info())
display(df_weather_HH)
```

#### Push the prepared data to a table in the database

```python
# Specify which table within your database you want to push your data to. Give your table an unambiguous name.
table_name = 'df_weather_HH_test_JB2'
schema = 'muc_analytics_21_1'
# If the specified table doesn't exist yet, it will be created
# With 'replace', your data will be replaced if the table already exists.
# This will take a minute or two...

# Write records stored in a dataframe to SQL database
if engine!=None:
    try:
        df_weather_HH.to_sql(name=table_name, # Name of SQL table
                        con=engine, # Engine or connection
                        if_exists='replace', # Drop the table before inserting new values 
                        schema=schema, # Use schmea that was defined earlier
                        index=False, # Write DataFrame index as a column
                        chunksize=5000, # Specify the number of rows in each batch to be written at a time
                        method='multi') # Pass multiple values in a single INSERT clause
        print(f"The {table_name} table was imported successfully.")
    # Error handling
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
        engine = None
```
